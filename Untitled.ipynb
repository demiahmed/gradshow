```python
for dirs in os.listdir("./pics"):
    try:
        key = list(df.loc[ df["name_full"] == dirs]["key"])[0]
        contents = os.listdir(f"./pics/{dirs}/Chosen Photo")
        file = [s for s in contents if "(edited)" in s][0]
        print(file)
        img = cv2.imread(f"./pics/{dirs}/Chosen Photo/{file}")
        h, w = img.shape[:2]
        aspect = w/h
        new_width = 512
        new_height = int(new_width/aspect)
        resized_img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_AREA)
        cropped_img = resized_img[0:512, 0:512]

        cv2.imwrite(f"./projects/2022/{key}/{key}.jpg", cropped_img, [cv2.IMWRITE_JPEG_QUALITY, 100])
    except Exception as e:
        print(dirs, e)
        continue
```
